```python
import pandas as pd

df = pd.read_csv("/content/Superstore.csv", encoding='latin1')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)

df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df['Year'] = df['Order Date'].dt.year

sales_by_category = df.groupby('Category')['Sales'].sum()
```

# Which product category generates the highest total sales?

```python
highest_total_sales= df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
```

```python
highest_total_sales
```

# Which sub-category is the least profitable despite high sales?

```python
subcat_analysis = df.groupby("Sub-Category")[["Sales", "Profit"]].sum()
```

```python
print(subcat_analysis.sort_values(by="Sales", ascending=False))
```

# 3. Which region contributes the highest profit?

```python
re_highest_pro=df.groupby("Region")["Profit"].sum().sort_values(ascending=False)
```

```python
print(re_highest_pro)
```

# 4. Which states are generating losses?

```python
State_losses=df.groupby("State")["Profit"].sum()
```

```python
loss_states=State_losses[State_losses<0]
```

```python
print(loss_states)
```

# 5. Which customer segment generates the highest average order value?

```python
highest_average_order_value = df.groupby("Segment")["Sales"].mean().sort_values(ascending=False)
```

```python
print(highest_average_order_value)
```

# **Discount & Profit Analysis**

# 6. Does giving higher discounts reduce profit?

```python
reduce_profit=df.groupby("Discount")["Profit"].mean().sort_values(ascending=False)
```

```python
print(reduce_profit)
```

# 8. Which products receive high discounts but still perform poorly?

```python
product_analysis = df.groupby("Product Name")[["Discount", "Profit"]].sum()

print(product_analysis.head())
```

```python
print(product_analysis.sort_values(by="Profit"))
```

# Which month had the highest sales?

```python
df["Month"] = df["Order Date"].dt.month_name()
```

```python
monthly_sales = df.groupby("Month")["Sales"].sum()

print(monthly_sales.sort_values(ascending=False))
```
